```python
import pandas as pd
import plotly.express as px

# Assuming 'data' is your DataFrame with the life expectancy data

# Create choropleth maps for male and female life expectancy
fig_male = px.choropleth(
    data_frame=df[df["Disaggregation"] == "male"],
    locations="Country Name",
    locationmode="country names",
    color="Value",
    color_continuous_scale="Viridis",
    title="Male Life Expectancy by Country",
    animation_frame="Year",
)

fig_female = px.choropleth(
    data_frame=df[df["Disaggregation"] == "female"],
    locations="Country Name",
    locationmode="country names",
    color="Value",
    color_continuous_scale="Viridis",
    title="Female Life Expectancy by Country",
    animation_frame="Year",
)

# Show the maps
fig_male.show()
fig_female.show()
```

## EDA

```python
filtered_data = df[(df['Year'] == 2020)][['Country Name', 'Year', 'Value']]
filtered_data
```

```python
sorted_data = filtered_data.sort_values(by='Value', ascending=False)
sorted_data
```

```python
df2020 = sorted_data = filtered_data.sort_values(by='Value', ascending=False)
df2020
```

```python
df2020.describe()
```

```python
# List of G20 member countries
g20_countries = [
    'Argentina', 'Australia', 'Brazil', 'Canada', 'China',
    'France', 'Germany', 'India', 'Indonesia', 'Italy',
    'Japan', 'Mexico', 'Russia', 'Saudi Arabia', 'South Africa',
    'South Korea', 'Turkey', 'United Kingdom', 'United States'
]
df2020[df2020["Country Name"].isin(g20_countries)]
```

```python
## The life expectency for United States is close to the mean
df2020[df2020["Country Name"].isin(g20_countries)].describe()
```

```python
df2020[df2020["Country Name"].isin(g20_countries)].sort_values(by="Value", ascending=True).head(10)
```

```python
# Calculate the median of the "Value" column
median_value = df2020["Value"].median()
filtered_countries = df2020[df2020["Country Name"].isin(g20_countries)]

# Sort the filtered data by the absolute difference from the median value
filtered_countries['Difference_from_Median'] = abs(filtered_countries['Value'] - median_value)
filtered_countries = filtered_countries.sort_values(by='Difference_from_Median', ascending=False)

# Select the top 10 countries closest to the median
result = filtered_countries.head(10)

# Drop the 'Difference_from_Median' column if you don't need it in the final result
result = result.drop(columns=['Difference_from_Median'])

# Print the result
print(result)
```

```python
import pandas as pd

# Assuming 'df2020' contains your DataFrame for the year 2020
# Calculate the median life expectancy in 2020
median_life_expectancy = df2020[df2020['Year'] == 2020]['Value'].median()

# Define a threshold for similarity, e.g., within 5% of the median
threshold = 0.05  # 5%

# Filter countries with life expectancy similar to the median
similar_countries = df2020[(df2020['Year'] == 2020) &
                           (abs(df2020['Value'] - median_life_expectancy) / median_life_expectancy <= threshold)]

# Print the result
print(similar_countries)
```
